```python
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import glorot_normal
from keras.callbacks import TensorBoard
from sklearn.model_selection import StratifiedKFold
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('12_features_data.csv')
labels=pd.read_csv('12_features_labels.csv')
```

```python
del data['Unnamed: 0']
data.head()
```

```python
del labels['Unnamed: 0']
labels.head()
```

```python
X=np.array(data)
Y=np.array(labels)
print(X.shape,Y.shape)
```

```python
def create_model():
    model=Sequential()
    model.add(Dense(10,activation='relu',kernel_initializer=glorot_normal(),input_dim=12))
    model.add(Dense(8,activation='relu',kernel_initializer=glorot_normal()))
    model.add(Dense(6,activation='relu',kernel_initializer=glorot_normal()))
    model.add(Dense(4,activation='relu',kernel_initializer=glorot_normal()))
    model.add(Dense(2,activation='relu',kernel_initializer=glorot_normal()))
    model.add(Dense(1,activation='sigmoid'))
    return model
```

```python
n_folds=5
i=0
accuracy,loss=[],[]
scores={}
skf = StratifiedKFold(n_folds)
for i, (train, test) in enumerate(skf.split(X, Y)):
        print("Running Fold", i+1, "/", n_folds)
        tensorboard=TensorBoard(log_dir='./logs/fold_{}'.format(i+1))
        x_train=np.array([X[i] for i in train])
        y_train=np.array([Y[i] for i in train])
        x_test=np.array([X[i] for i in test])
        y_test=np.array([Y[i] for i in test])
        model=create_model()
        model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
        model.fit(x_train,y_train,epochs=500,batch_size=32,callbacks=[tensorboard])
        result=model.evaluate(x_test,y_test,verbose=2)
        loss.append(result[0])
        accuracy.append(result[1])
scores['loss']=loss
scores['accuracy']=accuracy
testing_results=pd.DataFrame(scores)
```
